```python
import numpy as np
import matplotlib.pyplot as plt
from ml_algorithms.regressors import LinearRegression

NUM_SAMPLES = 100

X = np.arange(NUM_SAMPLES, dtype=np.float32)
a = 2.5
b = 100.3
y = a * X + b + np.random.randn(NUM_SAMPLES) * 10.

x = np.linspace(-10, 10, 100)

reg = LinearRegression()
reg.fit(X[..., np.newaxis], y[..., np.newaxis])

X_linspace = np.linspace(0, 100, 100)
y_linspace = np.linspace(0, 100, 100) * reg.W.item() + reg.B.item()

print(reg.W)
print(reg.B)

plt.scatter(X, y)
plt.plot(X_linspace, y_linspace, color="red")
plt.show()
```
