# Important obeservation
Eigenvectors multiplied by -1 are still a valid solution. In my implementation, the X axis is reversed in relation to sklearn, but this does not affect the validity of the result.

```python
from ml_algorithms.unsupervised import PCAWithScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA as SklearnPCA
from sklearn.preprocessing import StandardScaler

data = load_wine()
X = data.data
y = data.target

custom_pca = PCAWithScaler(n_components=2)
X_custom_pca = custom_pca.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
sklearn_pca = SklearnPCA(n_components=2)
X_sklearn_pca = sklearn_pca.fit_transform(X_scaled)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(X_custom_pca[:, 0], X_custom_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
axs[0].set_title('Custom PCAWithScaler')
axs[0].set_xlabel('PC1')
axs[0].set_ylabel('PC2')

axs[1].scatter(X_sklearn_pca[:, 0], X_sklearn_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
axs[1].set_title('sklearn PCA')

plt.tight_layout()
plt.show()
```
